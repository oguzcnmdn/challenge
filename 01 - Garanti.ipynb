```python
import pandas as pd
import numpy as np
import datetime
```

```python
df_edu = pd.read_csv(".\data\education.csv")
df_lan = pd.read_csv(".\data\languages.csv")
df_skills = pd.read_csv(".\data\skills.csv")
df_subs = pd.read_csv(".\data\submission.csv")
df_work_exp = pd.read_csv(".\data\work_experiences.csv")
test_users = pd.read_csv("test_users.csv")
train_users = pd.read_csv("train_users.csv")
```

```python
list_1 = df_work_exp["start_year_month"].tolist()
list_2 = []
for i in list_1:
    i_datetime = datetime.datetime.strptime(str(i),'%Y%m')
    list_2.append(i_datetime)
```

```python
df_work_exp["start_year_month_c"] = list_2
df_work_exp.drop("start_year_month", axis=1)
```

```python
from functools import reduce
dfs = [df_edu, df_lan, df_skills, df_subs, df_work_exp]
df_final = reduce(lambda left,right: pd.merge(left,right,on='user_id'), dfs)
```

```python
df_final_c = df_final.drop(columns=["start_year_month_x","end_year_month","degree","company_id","start_year_month_y"], axis=1)
```

```python
df_final_c.drop_duplicates(subset='user_id',keep='first', inplace=True)
df_final_c.sample(n=40)
```
